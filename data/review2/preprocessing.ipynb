```python
import pandas as pd
import numpy as np

def neg_sampling(N, path):

    concat_cond=[]
    for i in range(N):
        globals()['df_{}'.format(i)] = pd.read_csv(path.format(i), encoding="utf-8-sig")
        concat_cond.append(globals()['df_{}'.format(i)])
        
    # concatenation
    concat_df = pd.concat(concat_cond)
    # '일 전', '분 전', '시간 전', '일주일 전' , '어제' 제거
    concat_df = concat_df[~concat_df['날짜'].str.contains('전', na=False, case=False)]
    concat_df = concat_df[~concat_df['날짜'].str.contains('어제', na=False, case=False)]

    # negative review sampling
    negative_cond = (concat_df['맛'] < 4) | (concat_df['양'] < 4) | (concat_df['배달'] < 4)
    neg_df = concat_df.loc[negative_cond]
    
    # reset index
    concat_df = concat_df.reset_index()
    concat_df.pop('index')
    neg_df = neg_df.reset_index()
    neg_df.pop('index')
    
    
    return concat_df, neg_df
    

def saving_df(df, path):
    df.to_csv(path, encoding="utf-8-sig")
```
